### ***Data preparation***

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('/content/diabetes.csv')
pd.set_option('display.max_rows', None, 'display.max_columns', None)
df

```

```python
new_df = df.copy()
new_df.head(10)

```

```python
print(new_df.isnull().sum())
```

```python
new_df.info()
```

```python

for col in new_df.columns:
  mean_value = new_df[col].mean()
  new_df[col] = new_df[col].replace(0, mean_value)
  new_df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','Outcome']] = new_df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','Outcome']].astype(int)
new_df['BMI'] = new_df['BMI'].round(2)
new_df.head(10)
```

```python
new_df['Outcome'] = new_df['Outcome'].map({0: 'No Diabetes', 1: 'Diabetes'})
```

```python
new_df.describe()
```

```python
def classify__bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'

new_df['BMI_Category'] = new_df['BMI'].apply(classify__bmi)
print(new_df['BMI_Category'].value_counts())
new_df.head()
```

```python
new_df['Glucose_to_BMI'] = new_df['Glucose'] / new_df['BMI']
print(new_df[['Glucose', 'BMI', 'Glucose_to_BMI']].head())
```

```python
mean_glucose = np.mean(new_df['Glucose'])
std_glucose = np.std(new_df['Glucose'])
print(f"\n mean of Glucose {mean_glucose}")
print(f"\n std of Glucose: {std_glucose}")
```

```python
new_df['Glucose_Level'] = pd.cut(
    new_df['Glucose'],
    bins=[40, 100, 140, 200],
    labels=['Low', 'Medium', 'High']
)
print(new_df['Glucose_Level'].value_counts())
```

```python
new_df['Age_Category'] = pd.cut(
    new_df['Age'],
    bins=[0, 30, 50, 100],
    labels=['Young', 'Middle-Aged', 'Old']
)
print(new_df['Age_Category'].value_counts())
```

```python
new_df.tail()
```

***Data visualization ***

```python
plt.figure(figsize=(10, 8))
correlation_matrix = new_df[['BMI','Pregnancies','Glucose',	'BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Glucose distribution

plt.figure(figsize=(8, 5))
sns.histplot(new_df['Glucose'], kde=True, color='blue', bins=20)
plt.title('Distribution of Glucose Levels')
plt.xlabel('Glucose')
plt.ylabel('Frequency')
plt.show()
```

```python
# Relationship between glucose and BMI
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Glucose', y='BMI', data=new_df, hue='Outcome', palette='viridis')
plt.title('Relationship between Glucose and BMI')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()
```

```python
##Distribution of BMI Categories

#Percentage calculation
bmi_counts = new_df['BMI_Category'].value_counts()
bmi_percentages = (bmi_counts / bmi_counts.sum()) * 100

####
plt.figure(figsize=(8, 5))
plt.pie(bmi_percentages, labels=bmi_counts.index, autopct='%1.1f%%',shadow=True,startangle=90, colors=['#FCDFFF', '#D2B9D3', '#7E354D', '#571B7E'])
plt.title('Distribution of BMI Categories')
plt.legend( bbox_to_anchor=(1, 0.9))
plt.show()
```

```python
#Average Glucose Levels by Age Category
glucose_age_avg = new_df.groupby('Age_Category')['Glucose'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Age_Category', y='Glucose', data=glucose_age_avg, palette='viridis')
plt.title('Average Glucose Levels by Age Category')
plt.xlabel('Age Category')
plt.ylabel('Average Glucose Level')
plt.show()
```

```python
# Detect outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=new_df, palette='Set2')
plt.title('Boxplot to Detect Outliers in Numeric Columns')
plt.xticks(rotation=90)
plt.show()
```

```python
outcome_ = new_df['Outcome'].value_counts()
plt.figure(figsize=(6, 4))
plt.bar(outcome_.index, outcome_.values, color=['#EBDDE2','#C6AEC7'])
plt.title('Distribution of Outcome')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.pairplot(new_df, hue='Outcome', diag_kind='kde', palette='coolwarm')
plt.show()
```
